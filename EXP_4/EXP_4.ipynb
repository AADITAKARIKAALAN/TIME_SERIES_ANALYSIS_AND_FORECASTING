```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("taxi dataset.csv")
```

```python
df["datetime"] = pd.to_datetime(df[["year", "month", "day"]]) + pd.to_timedelta(df["hour_of_day"], unit='h')
```

```python
df.set_index("datetime", inplace=True)
```

```python
daily_avg = df["total_amount"].resample("D").mean()
```

```python
smoothed = daily_avg.rolling(window=7, center=True).mean()
```

```python
plt.figure(figsize=(14, 6))
plt.plot(daily_avg, label="Daily Average Total Amount", alpha=0.6)
plt.plot(smoothed, label="7-Day Rolling Average (Smoothed)", linewidth=2, color='orange')
plt.title("Trend Estimation and Smoothing of Total Amount")
plt.xlabel("Date")
plt.ylabel("Total Amount ($)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
